```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.groupby(['species'])[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].mean()
```

```python
# FINDINGS
# Setosa as significantly smaller petal length compared to versicolor and virginica.
# Virginica has the longest sepals and petals, followed by Versicolor, then Setosa.
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data = df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.boxplot(x="sepal_width", y="species", data=df)
plt.show()
```

```python
# Line Chart: Sepal Length vs Sample Index, grouped by species
plt.figure(figsize=(10, 5))
sns.lineplot(x=df.index, y="sepal_length", hue="species", data=df, marker="o")

plt.title("Sepal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend(title="Species")
plt.show()
```

```python
# Bar Chart: Average Petal Length per Species
plt.figure(figsize=(8, 5))
sns.barplot(x="species", y="petal_length", data=df, estimator=lambda x: x.mean())

plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["petal_length"], bins=15, kde=True, color="blue")

plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```
